```python
# add domain column
def domain(Website):
    if 'www.' in str(Website):
        return Website.split("www.")[1]
    else:
        return Website
df['domain'] = df.Website.apply(domain)
df.domain
```

```python
def prop_type(Company):
    if 'Bed & Breakfast' in Company:
        return 'Bed & Breakfast'
    elif 'Inn' in Company:
        return 'Inn'
    elif 'Hotel' in Company:
        return 'Hotel'
    elif 'Motel' in Company:
        return 'Motel'
    elif 'Bed and Breakfast' in Company:
        return 'Bed and Breakfast'
    elif 'bed and breakfast' in Company:
        return 'Bed and Breakfast'
    elif 'Resort' in Company:
        return 'Resort'
    elif 'B&B' in Company:
        return 'B&B'
    elif 'Lodge' in Company:
        return 'Lodge'
    else:
        return 'Homes'
df['prop_type'] = df.Company.apply(prop_type)
```

```python
df.prop_type.value_counts()
```

```python
df[df['prop_type'] == 'Homes'].Company.tolist()
```

```python
ver = df[df['Email Status'] == "Verified"]
len(ver.index)
```

```python
unver = df[~(df['Email Status'] == "Verified")]
len(unver.index)
```

```python
unver = unver[unver.Website.notnull()]
len(unver.index)
```
