```python
from selenium import webdriver
```

```python
options = webdriver.ChromeOptions()

# options.add_argument('headless')

# set the window size
options.add_argument('window-size=1200x600')

# initialize the driver
driver = webdriver.Chrome(chrome_options=options)

# https://www.linkedin.com/uas/login
# tooltip-dev-sale-price price-info
driver.get('https://www.buzzbuzzhome.com/ca/listview/lat_46.3255/lng_-96.1145/zoom_4/loc_Vancouver/cType_0,0,3/sType_0,0,0,4/rbType_1,1/lID_425/lType_2/page_2/seed_837')

driver.implicitly_wait(5)

# for elm in driver.find_elements_by_css_selector(".developer"):
#     print(elm.get_attribute('href'))
```

```python
builders = [];
for page in range(2):
    for elm in driver.find_elements_by_css_selector(".listview-item"):
        for el in elm.find_elements_by_css_selector(".tooltip-dev-sale-price"):
            if el.text == "Sales Pending":
                try:
                    elm2 = elm.find_element_by_css_selector(".dev-developer")
                    a_tag = elm2.find_element_by_tag_name('a')
                    builders.append(a_tag.get_attribute('href'))
                except:
                    print('nope')
    
    next_button = driver.find_element_by_id('pagerNext')
    next_button.click()

sales_pending = list(set(builders))

```
